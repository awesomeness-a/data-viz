# Vizualizing World Cup data with Seaborn

For this project, we're gonna be exploring data from the Fifa World Cup from 1930-2014.
This data is provided by <a href="https://www.kaggle.com/abecklas/fifa-world-cup">Kaggle</a>.

Using Seaborn, we'll create a series of plots that explore aggregates and distribution across the goals scored in World Cup games.

```python
# import the modules
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# load WorldCupMatches.csv into a DataFrame called df
df = pd.read_csv('WorldCupMatches.csv')
```

```python
# inspect the DataFrame using .head()
print(df.head())
```

```python
# the data has the goals scored in each match broken up by goals for the home team and goals for the away team. 
# We want to visualize the total number of goals scored in each match
df['Total Goals'] = df['Home Team Goals'] + df['Away Team Goals']
print(df.head())
```

```python
# create a bar chart visualizing how many goals were scored each year the World Cup was held between 1930-2014
sns.set_style('whitegrid')
sns.set_context('poster', font_scale=0.6)
f, ax = plt.subplots(figsize=(12, 7))
ax = sns.barplot(
    data=df,
    x='Year',
    y='Total Goals')
ax.set_title('Total number of goals in FIFA cup')
plt.show()
```

```python
# create a box plot to visualize the distribution of the goals data instead of just the average with a bar chart
df_goals = pd.read_csv('goals.csv')
print(df_goals.head())
```

```python
sns.set_context('notebook', font_scale=1.25)
f, ax2 = plt.subplots(figsize=(12, 7))
ax2 = sns.boxplot(
    data=df_goals,
    x='year',
    y='goals',
    palette='Spectral')
ax2.set_title('Goals distribution by Year')
plt.show()
```
